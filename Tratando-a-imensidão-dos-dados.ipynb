Adicionando os imports

````python
import pandas as pd
import numpy as np
from tabulate import tabulate
````

Carregar dados

````python
df_original = pd.read_csv('dados.csv', sep=';', engine='python')
````

Visualizar informações iniciais

````python
print("Informações gerais:")
print(df_original.info())
````

````python
print("\nPrimeiras 5 linhas:")
df_original.head()
````

````python
print("\nÚltimas 5 linhas:")
df_original.tail()
````

Limpeza da coluna 'Calories'

````python
df = df_original.copy()

df['Calories'] = df['Calories'].replace(r'^\s*$', np.nan, regex=True)
df['Calories'] = pd.to_numeric(df['Calories'], errors='coerce')
df['Calories'] = df['Calories'].fillna(0)

df
````

Limpeza da coluna 'Date'

````python
df['Date'] = df['Date'].replace(r'^\s*$', np.nan, regex=True)
df['Date'] = df['Date'].str.replace('"', '', regex=False)
df['Date'] = df['Date'].str.replace("'", '', regex=False)
df['Date'] = df['Date'].replace('20201226', '2020/12/26')
df['Date'] = df['Date'].fillna('1900/01/01')

df
````

Conversão da coluna Date

````python
df['Date'] = pd.to_datetime(df['Date'], format='%Y/%m/%d', errors='coerce')
df['Date'] = df['Date'].replace(pd.Timestamp('1900-01-01'), pd.NaT)

df
````

Remover valores nulos

````python
df = df.dropna()
df
````

Exibir tabela formatada com tabulate

````python
from IPython.display import Markdown

# Formatar tabela com `tabulate` e exibir como Markdown para melhor visualização no notebook
formatted_table = tabulate(df, headers='keys', tablefmt='grid', showindex=False)
display(Markdown(f'```\n{formatted_table}\n```'))
````
